```python
import random

def bubble_sort(arr):
    n = len(arr)
    swaps_count = 0

    for i in range(n - 1):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swaps_count += 1

    return arr, swaps_count

input_list = [random.randint(1, 100) for _ in range(10)]

print("Початковий список:", input_list)

sorted_list, swaps = bubble_sort(input_list.copy())

print("Відсортований список:", sorted_list)
print("Кількість перестановок:", swaps)
```
